## Introduction to Time Series Analysis

A **time series** is a sequence of data points collected at **regular time intervals**. Unlike typical datasets where observations are **independent**, time series data points are **ordered chronologically** and often **depend on past values**. Understanding **time series analysis** is crucial in many fields, as it helps in identifying patterns, making forecasts, and uncovering underlying structures in data.

Examples of time series data include:
- **Stock prices** (e.g., daily closing prices of Apple stock)
- **Epidemiological data** (e.g., COVID-19 infection rates over time)
- **Weather data** (e.g., daily temperature recordings)
- **Economic indicators** (e.g., monthly inflation rates)

Unlike traditional datasets where rows represent **independent observations**, time series exhibit **temporal dependencies**—meaning past values often influence future values. This makes time series analysis **unique** and requires specialized techniques.

The key challenges in time series analysis include:
- **Data dependencies** – Observations are not independent but influenced by previous values.  
- **Irregularities & missing data** – Time series data may have gaps or irregular intervals.  
- **Trend & seasonality effects** – Many real-world processes change over time due to external factors.

Pandas provides various **`dt.` accessor functions** to manipulate and extract information from datetime columns.

**Extracting Date & Time Components**
- `year` → Returns the year  
- `month` → Returns the month  
- `day` → Returns the day  
- `hour` → Returns the hour  
- `minute` → Returns the minute  
- `second` → Returns the second  
- `microsecond` / `nanosecond` → Returns microseconds / nanoseconds  

**Getting Names of Date Components**
- `day_name()` → Returns the weekday name (e.g., "Monday")  
- `month_name()` → Returns the month name (e.g., "January")  

**Weekdays & Calendar Information**
- `day_of_week` or `weekday` → Returns the weekday as a number (0=Monday, 6=Sunday)  
- `week` or `weekofyear` → Returns the week number of the year  
- `quarter` → Returns the quarter (1-4)  
- `isocalendar()` → Returns (Year, Week, Weekday) as a tuple  

**Special Calendar Information**
- `days_in_month` or `daysinmonth` → Number of days in the respective month  
- `is_leap_year` → Returns `True` if the year is a leap year  
- `is_month_end` / `is_month_start` → `True` if the date is at the start or end of the month  
- `is_quarter_end` / `is_quarter_start` → `True` if at the start or end of a quarter  
- `is_year_end` / `is_year_start` → `True` if at the start or end of the year  

**Conversions**
- `strftime(format)` → Formats a date as a string (`.strftime("%Y-%m-%d")`)  
- `to_period(freq)` → Converts a date into a **period** (`freq='M'` for month)  
- `to_timestamp()` → Converts a period back into an exact date  
- `tz_localize(tz)` → Sets a timezone (`tz='Europe/Berlin'`)  
- `tz_convert(tz)` → Converts a timezone (`'UTC' → 'Europe/Berlin'`)  

**Rounding & Adjustments**
- `floor(freq)` → Rounds down (e.g., `floor('D')` rounds to the start of the day)  
- `ceil(freq)` → Rounds up (e.g., `ceil('H')` rounds to the next full hour)  
- `round(freq)` → Rounds to the nearest unit (e.g., `round('T')` rounds to the nearest minute)

### Key Components of Time Series Data
Time series data often exhibit **three fundamental patterns**:  

1. **Trend**

A **trend** represents the long-term movement of a time series, showing an **increasing** or **decreasing** pattern over time.  
- Example: The **global average temperature** has shown an **upward trend** over the past century due to climate change.
- Trends can be **linear or non-linear** and can shift due to external factors.

**Detecting a trend**:  
- **Rolling mean smoothing** helps visualize trends.
- **Differencing** (subtracting the previous value from the current one) helps remove trends for better analysis.
 

```python
import pandas as pd

df = pd.read_csv("weatherdata_cleaned.csv")
df["date"] = pd.to_datetime(df["date"])
df
```

```python
import matplotlib.pyplot as plt

# Plot the raw time series
plt.figure(figsize=(12, 5))
plt.plot(df["date"], df["mean_temperature"], label="Mean Temperature", color="blue")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Raw Time Series: Mean Temperature Over Time")
plt.legend()
plt.show()
```

```python
df["rolling_avg"] = df["mean_temperature"].rolling(window=3, min_periods=1).mean()
df.head()
```

```python
# Plot the original data and the trend
plt.figure(figsize=(12, 5))
plt.plot(df["date"], df["mean_temperature"], label="Mean Temperature (Raw Data)", alpha=0.5)
plt.plot(df["date"], df["rolling_avg"], label="365-Day Moving Average (Trend)", color="red", linewidth=2)
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Trend Analysis: Moving Average Over Time")
plt.legend()
plt.show()
```

```python
# Define maximum temperature by year
df["year"] = df["date"].dt.year
yearly_mean_temp = df.groupby("year")["mean_temperature"].mean().reset_index()

# Create plot
plt.figure(figsize=(12, 6))
plt.plot(yearly_mean_temp["year"][1:], yearly_mean_temp["mean_temperature"][1:], marker="o", linestyle="-", label="Mean Temperature per Year")
plt.xlabel("Year")
plt.ylabel("Mean Temperature (°C)")
plt.title("Yearly Mean Temperature")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
yearly_mean_temp["rolling_avg"] = yearly_mean_temp["mean_temperature"].rolling(window=5, min_periods=1).mean()
yearly_mean_temp.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(yearly_mean_temp["year"][1:], yearly_mean_temp["mean_temperature"][1:], marker="o", linestyle="-", label="Mean Temperature per Year")
plt.plot(yearly_mean_temp["year"][1:], yearly_mean_temp["rolling_avg"][1:], label="5-Year Moving Average (Trend)", color="red", linewidth=2)
plt.xlabel("Year")
plt.ylabel("Mean Temperature (°C)")
plt.title("Yearly Mean Temperature")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

**Differencing** is a technique **to remove trends** and make a time series stationary (i.e., with constant mean and variance).
- First-order differencing removes linear trends.
- Second-order differencing removes non-linear trends.

```python
yearly_mean_temp["1st Order Difference"] = yearly_mean_temp["mean_temperature"].diff()

# Plot original data vs. differenced data
plt.figure(figsize=(12, 5))
plt.plot(yearly_mean_temp["year"], yearly_mean_temp["mean_temperature"], label="Mean Temperature (Raw Data)", alpha=0.5)
plt.plot(yearly_mean_temp["year"], yearly_mean_temp["1st Order Difference"], label="First-Order Differencing", color="purple", linestyle="dashed")
plt.xlabel("Date")
plt.ylabel("Temperature Change (°C)")
plt.title("First-Order Differencing to Remove Trend")
plt.legend()
plt.show()
```

2. **Seasonality**

**Seasonality** refers to **recurring patterns** in time series data that occur at regular intervals (e.g., daily, weekly, yearly). Seasonality arises when a time series exhibits **cyclical behavior** due to external factors such as climate, holidays, or market cycles. Ignoring seasonality can lead to poor forecasting performance.

**Examples of Seasonality in Different Fields**:
- **Retail & Sales**️: Higher sales in December due to holiday shopping.
- **Weather Patterns**️: Increased rainfall during specific seasons.
- **Epidemiology**: Flu infections peaking in winter months.
- **Energy Consumption**: Higher electricity usage in summer due to air conditioning.

**Detecting Seasonality**:
To identify seasonality in a time series, we can:
- **Resample data to different time intervals** (e.g., weekly, monthly).  
- **Use rolling statistics to smooth fluctuations**.  
- **Apply Fourier Transform or Seasonal Decomposition (STL)**. 



```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform Seasonal Decomposition (using additive model)
decomposition = seasonal_decompose(df["mean_temperature"], model="additive", period=365)
decomposition
```

The seasonal decomposition technique breaks a time series into three main components:
1. Trend (Red Line) – The long-term increase or decrease in mean temperature.
2. Seasonality (Green Line) – Recurring yearly temperature patterns (e.g., warmer summers, colder winters).
3. Residual (Gray Line) – The unexplained variation (random fluctuations or anomalies).

```python
fig, axes = plt.subplots(4, 1, figsize=(12, 8), sharex=True)

# Original Data
axes[0].plot(df["date"], df["mean_temperature"], label="Original Data", color="blue")
axes[0].legend()
axes[0].set_title("Original Data")

# Trend Component
axes[1].plot(df["date"], decomposition.trend, label="Trend", color="red")
axes[1].legend()
axes[1].set_title("Trend")

# Seasonal Component
axes[2].plot(df["date"], decomposition.seasonal, label="Seasonality", color="green")
axes[2].legend()
axes[2].set_title("Seasonality")

# Residual Component
axes[3].plot(df["date"], decomposition.resid, label="Residual (Noise)", color="gray")
axes[3].legend()
axes[3].set_title("Residual (Noise)")

# Global Title & Layout Adjustment
fig.suptitle("Seasonal Decomposition of Mean Temperature", fontsize=14)
plt.xlabel("Date")
plt.tight_layout()
plt.show()
```

Not every time series has **equally strong seasonality**.  
A simple way to measure the strength of seasonality is the **ratio of seasonal variance to total variance**:

$
\text{Seasonal Strength} = 1 - \frac{\text{Variance of Residuals}}{\text{Variance of Original Series}}
$

- **Close to 1** → High seasonality (strongly recurring patterns) 
- **Close to 0** → Weak or no seasonality  

```python
seasonal_strength = 1 - (decomposition.resid.var() / df["mean_temperature"].var())
print(seasonal_strength)
```

3. **Autocorrelation**

**Autocorrelation** measures how a time series correlates with **its own past values**. It indicates whether past observations influence future values, making it a fundamental concept in time series analysis.

**Why Does Autocorrelation Matter?**
- **High autocorrelation** means that previous values strongly affect future values.
- **Low or no autocorrelation** suggests a more random pattern.
- **Detecting autocorrelation** helps in choosing the right forecasting model.

**Examples of Autocorrelation in Real-World Data**
- **Stock Prices**: If a stock price is high today, it is more likely to be high tomorrow.
- **Temperature Data️**: Today’s temperature is often similar to yesterday’s.
- **Epidemiology**: COVID-19 case numbers often depend on previous days' cases.
- **Energy Consumption**: Power usage patterns show strong autocorrelation due to daily routines.

**Detecting Autocorrelation**
Autocorrelation is typically analyzed using:
- **Autocorrelation Function (ACF)** – Measures how correlated a time series is with itself at different time lags.  
- **Partial Autocorrelation Function (PACF)** – Identifies the most relevant time lags affecting the series.  
- **Lag Plots** – Scatterplots comparing a time series with its lagged values. 

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot Autocorrelation Function (ACF)
plt.figure(figsize=(14, 8))
plot_acf(df["mean_temperature"], lags=50)  # 50 lags to check short-term correlation
plt.title("Autocorrelation Function (ACF) of Mean Temperature")
plt.show()
```

The **autocorrelation function (ACF)** measures how strongly a current value correlates with its previous values (**lags**). It therefore indicates whether past values in the time series can help predict future values.
- A high ACF value for a particular lag means that the current value of the time series correlates strongly with the value from the past.
- The ACF can fall slowly if there is a trend in the time series.
- If a seasonal pattern exists, the ACF shows a periodic structure (e.g. high correlation for lags that correspond to the seasonal cycle).

Example: Suppose we have a daily temperature time series. If the temperature of one day is strongly related to the temperature of the previous day, we expect a high autocorrelation at lag 1. If the pattern is repeated weekly, we also see high values at lag 7, 14, etc.

```python
df["diff_1"] = df["mean_temperature"].diff()
plot_acf(df["diff_1"].dropna(), lags=50)
plt.title("ACF after First-Order Differencing")
plt.show()
```

```python
# Plot Partial Autocorrelation Function (PACF)
plt.figure(figsize=(12, 5))
plot_pacf(df["mean_temperature"], lags=50)  # 50 lags to check direct effects
plt.title("Partial Autocorrelation Function (PACF) of Mean Temperature")
plt.show()
```

The **partial autocorrelation (PACF)** measures the direct relationship between a current value and a delayed value (lag) after the influence of all intervening lags has been removed.
While the ACF measures the overall relationship between a current observation and all its previous values, the PACF only measures the direct relationship between an observation and a specific lag. It ignores indirect effects.

```python
df["seasonal_diff"] = df["mean_temperature"].diff(periods=365)
plot_acf(df["seasonal_diff"].dropna(), lags=50)
plt.title("ACF after Seasonal Differencing")
plt.show()
```

```python
df = df.sort_values(by='date')
df.set_index('date', inplace=True)
```

```python
# Specify exogene variables
exog_vars = ['mean_relative_humidity', 'mean_wind_speed', 'mean_pressure', 'precipitation_height']
df_exog = df[exog_vars].interpolate().fillna(df[exog_vars].mean())
```

```python
# Subset smaller dataset
df_subset = df.iloc[-1000:]
df_exog_subset = df_exog.iloc[-1000:]
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

sarimax_model = SARIMAX(df_subset['mean_temperature'], 
                        exog=df_exog_subset, 
                        order=(1, 1, 1),  
                        # (p = # of autoregressive terms, i.e. how much past values influence the future,
                        #  d = # of differences to make the series stationary,
                        #  q = # of moving average terms, i.e. how much past error terms influence the forecast)
                        seasonal_order=(1, 1, 0, 12),
                        # (P = Seasonal autoregressive term,
                        #  D = Seasonal difference, i.e. removes seasonal trends,
                        #  Q = Seasonal moving average term, set to 0 means no seasonal MA,
                        #  S = Seasonal period, e.g. 12 months for yearly seasonality)
                        enforce_stationarity=False, # allows non-stationary models if they perform better
                        enforce_invertibility=False) # prevents the model from forcing invertibility

model_fit = sarimax_model.fit(disp=False)
```

```python
# Define forecast period
forecast_steps = 20

# Exogene variables for forecast (repeat last value)
last_exog_values = df_exog_subset.iloc[-1].values
future_exog = pd.DataFrame([last_exog_values] * forecast_steps, columns=df_exog_subset.columns)

# generate forecast
forecast = model_fit.forecast(steps=forecast_steps, exog=future_exog)

# Create time axis for forecast
forecast_dates = pd.date_range(start=df_subset.index[-1], periods=forecast_steps + 1, freq='D')[1:]
```
